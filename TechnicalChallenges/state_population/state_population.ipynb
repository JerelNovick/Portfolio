<p>Given the following datasets on <a href ='https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_populations.csv'>state population</a> and <a href='https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_births.csv'>state deaths and births</a>, <b>create a list that has each state's population growth rate, sorted in descending order.</b> Note you will need to join the two datasets.</p>

```python
import numpy as np
import pandas as pd
```

```python
population = pd.read_csv("https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_populations.csv") 
birth_death = pd.read_csv("https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/us_state_births.csv") 
```

```python
population.head()
```

```python
birth_death.head()
```

```python
df = pd.merge(population, birth_death, on='State')
df.head()
```

```python
#add net change column
df['Net Change'] = df['Births'] - df['Deaths']

#add growth rate column
df['Growth Rate']=df['Net Change'] *100/df['Population']
df.head()
```

```python
#sort dataframe
df = df.sort_values(by='Growth Rate', ascending=False).reset_index(drop=True)
df.head()
```
