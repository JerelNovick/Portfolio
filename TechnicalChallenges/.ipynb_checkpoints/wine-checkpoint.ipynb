```python
#Importing packages.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
import datetime
import pandas_profiling as pp
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import time
%matplotlib inline
```

```python
#reading in data
data = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/winequality-red.csv',sep = ';') 
data.head()
```

```python
pp.ProfileReport(data)
```

```python
### Setting up for numerical encoding ####
# this code is pretty much plug and go, since we don't need an unique identifier
# lets just leave this blank
unique_id = ['']

# then we're going to specify our target (e.g. what we want the decision tree to 
# optimize on)
target_col = ['quality']

# we don't have categorical columns
# lets put the cutoff to 0
category_col   = data.nunique()[data.nunique() < 0].keys().tolist()
category_col   = [x for x in category_col if x not in target_col]

# now we're going to pull all the numerical columns 
# the logic here is if it's not a category, target, or userid, it's numeric
# we'll be normalizing these columns
numeric_cols   = [x for x in data.columns if x not in category_col + target_col + unique_id]

# next we want to break apart binary columns 
# binary column values will live in 1 column
binary_cols   = data.nunique()[data.nunique() == 2].keys().tolist()

# lastly we want to specify all non_binary columns
# non binary numeric columns will be broken apart into different columns so we can
# interpret them as binary
nonbinary_cols = [i for i in category_col if i not in binary_cols]



### Numerical encoding ####
#We're going to use the LabelEncoder to encode binary columns
le = LabelEncoder()
for i in binary_cols :
    data[i] = le.fit_transform(data[i])

# Duplicating columns for multi-value columns
data = pd.get_dummies(data = data, columns = nonbinary_cols )

# Scaling Numerical columns
scaled = StandardScaler().fit_transform(data[numeric_cols])
#creating a dataframe for the output
scaled = pd.DataFrame(scaled, columns=numeric_cols)

# Last we're going to clean things up
# We're going to drop the original values and merge the scaled values 
# and binary values
df_data_og = data.copy()
data = data.drop(columns = numeric_cols,axis = 1)
data = data.merge(scaled,left_index=True,right_index=True,how = 'left')

#viewing our output!
data.head()
```
