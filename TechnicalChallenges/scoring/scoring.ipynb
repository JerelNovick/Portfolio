<p>There are a few ways we can score in American Football:
  <ul>
  <li> 1 point - After scoring a touchdown, the team can choose to score a field goal</li>
  <li> 2 points - (1) after scoring touchdown, a team can choose to score a conversion, when the team attempts to score a secondary touchdown or (2) an uncommon way to score, a safety is score when the opposing team causes the ball to become dead</li>
  <li> 3 points - If no touchdown is scored on the possession, a team can attempt to make a field goal</li>
  <li> 6 points - Awarded for a touchdown</li>
  </ul>
  <br> Given the above the potential point values for American Football are:
  <ul>
  <li> 2 points - safety
  <li> 3 points - only field goal
  <li> 6 points - only touchdown
  <li> 7 points - touchdown + field goal
  <li> 8 points - touchdown + conversion
  </ul>
  </p>
  <br>
  <p>Given a score value, can you <b>write a function that lists the possible ways the score could have been achieved?</b>
  <br>For example, if you're given the value 10, the potential values are:
  <ul>
  <li> 8 points (touchdown + conversion) + 2 points (safety)
  <li> 6 points (only touchdown) + 2x2 points (safety)
  <li> 7 points (touchdown + field goal) + 3 points (only field goal)
  <li> 5x2 points (safety)
  <li> 2x2 points (safety) + 2x3 points (only field goal)
  </ul>
  </p>

```python
import copy
```

```python
#create a list of point values
potential_values = [2,3,6,7,8]
table = {0:[{}]}
```

```python
def possible_scores(score):
    #create empty table
    table = {0:[{}]}
    for i in range (1, score +1):
        table[i] = []
        for v in potential_values:
            if i - v <0 or i - v not in table:
                #if the number is negative or not in point_value, we will skip it
                continue
            for score_table in table[i - v]:
                #append each new score
                new_score_table = copy.copy(score_table)
                new_score_table[v] = new_score_table.get(v,0) + 1
                
                if new_score_table not in table[i]:
                    #if the arrangement is not already in the table then add it
                    table[i].append(new_score_table)
    return table[score]
```
