<p>Suppose you are given the following dataframe containing food, servings per container, calories per serving, and price per container. You'll notice that each food's price, serving size, serving per container widely varies. <b>Create a sorted list (lowest to highest) with the food and the associated price per calorie. </b></p>


```python
#code to generate example dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/snackfood.csv')
df.head()
```

```python
#create calories per container column
df['calories_per_container'] = df['servings_per_container'] * df['calories_per_serving']
df.head()
```

```python
#create price per calorie column
df['price_per_calorie'] = df['price_per_container'] / df['calories_per_container']
df.head()
```

```python
#sort by price per calorie
df2=df[['food', 'price_per_calorie']].sort_values(by='price_per_calorie', ascending=True).reset_index(drop=True)
df2
```
